```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/drgeary/datasets/main/Percent_Change_in_Consumer_Spending__2020-2022.csv')

df
```

```python
df.info()
```

```python
df.columns.tolist()
```

```python
fig, ax=plt.subplots()
g = df['Date']
h = df['All merchant category codes spending']
ax.set_xlabel('Date')
ax.set_ylabel('All merchant category codes spending')
ax.set_title('All merchant category codes spending vs Date')
ax.scatter(g, h, color = 'green', alpha=0.5)
plt.show()

#This scatter plot shows 2 drastic shifts in the data starting date is Jan 2020 to Mar 2023
```
